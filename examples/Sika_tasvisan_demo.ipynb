```python
import matplotlib.pyplot as plt
import numpy as np
exp1=sika.sika_expconfig(ef=5)
exp1.sample.a     = 5
exp1.sample.b     = 5
exp1.sample.c     = 12
exp1.sample.alpha = 90
exp1.sample.beta  = 90
exp1.sample.gamma = 120
#exp1.sample.mosaic  = 30
#exp1.sample.vmosaic = 30
exp1.sample.u       = np.array([1, 0, 0])
exp1.sample.v       = np.array([0, 0, 1])



scanno= 399  
df=sika.sika_data_to_pd(path, scanno)
#print(df)
hklw1=df[['qh', 'qk', 'ql', 'en','detector']]
plt.plot(hklw1['en'], hklw1['detector'], 'bo')
```

```python

# These are parameters:        
# en1, en2, ratio, w1, w2, int, bg, temp
# In this case, the scan is a energy scan. it is able to fit two peaks
# The temperature should be fixed. 
# The sqw function is the default one: SqwDemo and PrefDemo
# parameter order: en1, en2, ratio, gamma1, gamma2, intensity, background, temperature
#initialguess = {'en1': 1.55, 'en2': 3.4, 'ratio':  2.6,  'gamma1': 0.1,  'gammma2': 0.09, 'int': 7300,  'bg': 2,  'temp':1.5}

initialguess = [1.55, 3.4, 2.6, 0.10, 0.09, 7300, 10, 1.5]
fixpar       = [1,      1,   1,    1,    1,    1,   1,   0]

finalpar, fitteddata = sika.sika_convfit(hklw=hklw1, exp = exp1,  initial=initialguess, fixlist=fixpar, magion= "Co2")
plt.plot(hklw1['en'], hklw1['detector'], 'bo')  #raw data
plt.plot(fitteddata[:,3],fitteddata[:,4], 'r-') #fitted curve
```
